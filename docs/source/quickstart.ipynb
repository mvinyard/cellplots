## Load some data

We'll use single-cell data since we have some built in functions for `umap`.

```python
import scanpy as sc
```

```python
adata = sc.datasets.pbmc3k_processed()
print(adata)
```

### UMAP

```python
axes = cp.umap(adata)
```

```python
cmap = {
    "CD4 T cells": "#448aff",
    "B cells": "#1565c0",
    "CD14+ Monocytes": "#009688",
    "NK cells": "#8bc34a",
    "CD8 T cells": "#ffc107",
    "FCGR3A+ Monocytes": "#ff9800",
    "Dendritic cells": "#f44336",
    "Megakaryocytes": "#ad1457",
}
```

```python
axes = cp.umap(adata, groupby="louvain", cmap=cmap)
```

```python
axes = cp.umap_manifold(adata, groupby="louvain", c_background=cmap)
```
